```python
#Let's keep only those countries in the dataset;

countries = ['United States', 'Brazil', 'United Kingdom', 'Italy', 'France', 'Spain', 'Mexico', 'India', 'Iran', 'Peru', 
             'Taiwan', 'New Zealand', 'Singapore', 'Australia', 'South Korea', 'Vietnam', 'Thailand', 'Nigeria', 'Ethiopia',
             'Japan']
countries_ox = ox['CountryName'].unique()
delete = [country for country in countries_ox if country not in countries]

ox_20 = ox

for country in delete:
    ox_20.drop(ox_20[ox_20['CountryName'] == country ].index , inplace=True)
    
ox_20.shape

```

```python
ox_20_cases = ox_20[['CountryName', 'Date', 'ConfirmedCases', 'ConfirmedDeaths']].groupby(by=['CountryName', 'Date'], as_index=False).sum()
ox_20_cases
```

```python
cases_per_country = {}

for country in countries:
    cases_per_country[country] = list(ox_20_cases[ox_20_cases['CountryName'] == country]['ConfirmedCases'])
```

```python
print(cases_per_country)
```

```python
days_no_cases = {}

for country in cases_per_country.keys():
    days = 0
    for cases in cases_per_country[country]:
        if cases == 0:
            days += 1
        else:
            break
    days_no_cases[country] = days

days_no_cases
#Days in 2020 before the first confirmed case for each country.
```

```python
#Let's create a function to count the number of days before an event occurs.

ox_20 #US and UK are separated by region. Let's groupby country and take the average value of each category per date.

ox_country = ox_20.groupby(by=['CountryName', 'Date'], as_index=False).agg({'C1_School closing':'mean',
                                                                            'C2_Workplace closing':'mean', 
                                                                            'C3_Cancel public events':'mean',
                                                                            'C4_Restrictions on gatherings':'mean', 
                                                                            'C5_Close public transport':'mean',
                                                                            'C6_Stay at home requirements':'mean', 
                                                                            'C7_Restrictions on internal movement':'mean',
                                                                            'C8_International travel controls':'mean', 
                                                                            'E1_Income support':'mean',
                                                                            'E2_Debt/contract relief':'mean', 
                                                                            'E3_Fiscal measures':'mean',
                                                                            'E4_International support':'mean', 
                                                                            'H1_Public information campaigns':'mean',
                                                                            'H2_Testing policy':'mean', 
                                                                            'H3_Contact tracing':'mean',
                                                                            'H4_Emergency investment in healthcare':'mean',
                                                                            'H5_Investment in vaccines':'mean',
                                                                            'ConfirmedCases':'sum', 
                                                                            'ConfirmedDeaths':'sum', 
                                                                            'StringencyIndexForDisplay':'mean',
                                                                            'StringencyLegacyIndexForDisplay':'mean', 
                                                                            'GovernmentResponseIndexForDisplay':'mean',
                                                                            'ContainmentHealthIndexForDisplay':'mean', 
                                                                            'EconomicSupportIndexForDisplay':'mean'})

def list_per_country(empty_dict, column_name):
    
    for country in countries:
        empty_dict[country] = list(ox_country[ox_country['CountryName'] == country][column_name])
    return empty_dict
    
def count_days(new_dict, prev_dict):
    
    for country in prev_dict.keys():
        days = 0
        for value in prev_dict[country]:
            if value == 0:
                days += 1
            else:
                break
        new_dict[country] = days
    return new_dict
```

# Create a Global Database for statistical Analysis

In this next part, I want to create a database that group information from different sources, group them per country and perform statistical analysis to determine the efficiency of the political response of each country and look for any significant pattern.

```python
#Import all the sources:

oxford = pd.read_csv('C:\\Users\\sarah\\Documents\\IRONHACK\\Final_Project_Mod2\\data\\oxford_clean.csv', index_col=0)
owid = pd.read_csv('C:\\Users\\sarah\\Documents\\IRONHACK\\Final_Project_Mod2\\owid-covid-data.csv', index_col=0)
```

```python
#Group all the sources per country. We should have only one line per country, with wether the mean, sum, max.. per column.
#1 is Oxford DataBase
oxford = oxford[['CountryName', 'RegionName', 'Date', 'E3_Fiscal measures', 'E4_International support',
                 'H4_Emergency investment in healthcare', 'H5_Investment in vaccines', 'ConfirmedCases', 
                 'ConfirmedDeaths', 'StringencyIndexForDisplay', 
                 'StringencyLegacyIndexForDisplay', 'GovernmentResponseIndexForDisplay',
                 'ContainmentHealthIndexForDisplay', 'EconomicSupportIndexForDisplay']]

oxford = oxford.groupby(by=['CountryName', 'Date'], as_index=False).agg({'E3_Fiscal measures':'sum',
                                                                         'E4_International support':'sum', 
                                                                         'H4_Emergency investment in healthcare':'sum',
                                                                         'H5_Investment in vaccines':'sum',
                                                                         'ConfirmedCases':'sum', 
                                                                         'ConfirmedDeaths':'sum', 
                                                                         'StringencyIndexForDisplay':'mean',
                                                                         'StringencyLegacyIndexForDisplay':'mean', 
                                                                         'GovernmentResponseIndexForDisplay':'mean',
                                                                         'ContainmentHealthIndexForDisplay':'mean', 
                                                                         'EconomicSupportIndexForDisplay':'mean'})

oxford = oxford.groupby(by=['CountryName'], as_index=False).agg({'E3_Fiscal measures':'max',
                                                                 'E4_International support':'max', 
                                                                 'H4_Emergency investment in healthcare':'max',
                                                                 'H5_Investment in vaccines':'max',
                                                                 'ConfirmedCases':'max', 
                                                                 'ConfirmedDeaths':'max', 
                                                                 'StringencyIndexForDisplay':'mean',
                                                                 'StringencyLegacyIndexForDisplay':'mean', 
                                                                 'GovernmentResponseIndexForDisplay':'mean',
                                                                 'ContainmentHealthIndexForDisplay':'mean', 
                                                                 'EconomicSupportIndexForDisplay':'mean'})
#max values for investment and number of cases, because cumulative values in sources
#mean for indexes, to take into account the number of days without response

#2 is Our World In Data (owid)
owid = owid[['location', 'total_tests', 'total_tests_per_thousand', 'tests_per_case', 
             'tests_units', 'population', 'population_density', 'median_age', 'aged_70_older', 'gdp_per_capita', 
             'extreme_poverty', 'handwashing_facilities', 'hospital_beds_per_thousand', 'life_expectancy']]
owid = owid.fillna(0)
owid = owid.groupby(by=['location'], as_index=False).agg('max')
```

```python
#Before merging, get the same number of rows, with the same countries

owid_countries = set(owid['location'].unique())
ox_countries = set(oxford['CountryName'].unique())
keep = list(owid_countries.intersection(ox_countries))

del_owid = [country for country in list(owid_countries) if country not in keep]
del_ox = [country for country in list(ox_countries) if country not in keep]

for country in del_owid:
    owid.drop(owid[owid['location'] == country].index , inplace=True)
for country in del_ox:
    oxford.drop(oxford[oxford['CountryName'] == country].index , inplace=True)
    
print(owid.shape)
print(oxford.shape)
```

```python
#Merging the 2 databases
covid = oxford.merge(owid, left_on='CountryName', right_on='location')
covid.drop('location', axis=1, inplace=True)
covid
```
